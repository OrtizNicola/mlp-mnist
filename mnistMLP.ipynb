## Using tensorflow to tackle the MNIST dataset

This is an implementation of a tensorflow MLP to predict the MNIST digits. The training will involve hyperparameter tuning with keras tuner and monitoring of the model using TensorBoard.

```python
import tensorflow as tf
import tensorflow.keras.callbacks as cb
import tensorflow.keras.layers as lyrs
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# load the data
data = tf.keras.datasets.mnist.load_data()
```

```python
# split and scale the data
(xTrainFull, yTrainFull), (xTest, yTest) = data
xTrainFull, xTest = xTrainFull / 255, xTest / 255 
```

```python
# get the validation set
xTrain, xVal = xTrainFull[5000:], xTrainFull[:5000]
yTrain, yVal = yTrainFull[5000:], yTrainFull[:5000]
```

```python
# plot some instances
def plotImages(x):
    nDigits = x.shape[0]
    nCols = 5
    nRows = int(nDigits / 5 + 1)
    plt.figure(figsize=(2*nCols, 2*nRows))
    for i in range(nDigits):
        plt.subplot(nRows, nCols, i + 1)
        plt.imshow(x[i], cmap="binary")
        plt.tick_params(left=False, labelleft=False,
                        bottom=False, labelbottom=False)
    plt.subplots_adjust(wspace=0,hspace=0)
    plt.show()
plotImages(xTrain[:20])
```

To get an appropiate learning rate, it's useful to increase it at each iteration and see when the loss diverges. This callback let us do just that:

```python
class ExpLearningRate(cb.Callback):
    def __init__(self, factor):
        self.factor = factor
        self.rates = []
        self.lossi = []
    def on_batch_end(self, batch, logs=None):
        lr = self.model.optimizer.learning_rate.numpy() * (1 + self.factor / 100)
        self.model.optimizer.learning_rate = lr
        self.rates.append(lr)
        self.lossi.append(logs["loss"])
```

```python
tf.keras.backend.clear_session()
tf.random.set_seed(42)
```

```python
model = tf.keras.Sequential([
    lyrs.Flatten(input_shape=[28, 28]),
    lyrs.Dense(256, activation="relu"), 
    lyrs.Dense(256, activation="relu"),
    lyrs.Dense(10, activation="softmax")
])
```

```python
optimizer = tf.keras.optimizers.SGD(learning_rate=1e-4)
model.compile(loss="sparse_categorical_crossentropy", 
              optimizer=optimizer,
              metrics=["accuracy"])
expLR = ExpLearningRate(factor=1) # grow the lr 1% each iteration
```

```python
history = model.fit(xTrain, yTrain, epochs=1, # 1 epoch is enough for the loss to diverge
                   validation_data=(xVal, yVal),
                   callbacks=[expLR])
```

```python
plt.plot(expLR.rates, expLR.lossi)
plt.gca().set_xscale('log')
plt.hlines([min(expLR.lossi)], min(expLR.rates), max(expLR.rates), "r")
plt.axis([min(expLR.rates), max(expLR.rates), 0, expLR.lossi[0]])
plt.grid()
plt.xlabel("Learning rate")
plt.ylabel("Loss");
```

It looks like this problem allows us to use a big learning rate. The loss grows too much when the lr is close to 1 so we can try going with 0.5:

```python
# use this to save the results of each run
nRun = 0
```

```python
tf.keras.backend.clear_session()
tf.random.set_seed(42)

model = tf.keras.Sequential([
    lyrs.Flatten(input_shape=[28, 28]),
    lyrs.Dense(256, activation="relu"), 
    lyrs.Dense(256, activation="relu"),
    lyrs.Dense(10, activation="softmax")
])

optimizer = tf.keras.optimizers.SGD(learning_rate=5e-1)
model.compile(loss="sparse_categorical_crossentropy", 
              optimizer=optimizer,
              metrics=["accuracy"])

# save the logs 
nRun += 1
logDir = Path() / "mnist-logs" / f"run-{nRun:03d}"

# callbacks
earlyStopping = cb.EarlyStopping(patience=20)
checkpoint = cb.ModelCheckpoint("mnist-model.keras", save_best_only=True)
tensorBoard = cb.TensorBoard(logDir)

history = model.fit(xTrain, yTrain, epochs=200, 
                   validation_data=(xVal, yVal),
                   callbacks=[earlyStopping, checkpoint, tensorBoard])
```

```python
# rollback to the best model
model = tf.keras.models.load_model("mnist-model.keras")
model.evaluate(xTest, yTest)
```
